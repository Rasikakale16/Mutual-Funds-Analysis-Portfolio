```python
def calculator(investment, r, years):
   #returns calculator
    months = years *12
    
    #compound interest formula
    i = r/100/12
    
    #calculation for FV(Future value)
    a = (1 + i)**months-1

    b = (1+i)/i
    
    #final FV calculation
    FV = investment * a * b
    #print(FV)
    
    return FV
    
```
